```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("amsterdam_weekdays.csv")
```

```python
df = pd.read_csv('amsterdam_weekdays.csv')
```

```python
df.to_excel('amsterdam_weekdays.xlsx', index=False)
```

```python
df = pd.read_excel('amsterdam_weekdays.xlsx')
```

```python
print("Первые 5 строк данных:")
print(df.head())
```

```python
max_rating_realSum = df[df['realSum'] == df['realSum'].max()]
print("Сумма с наивысшим рейтингом:")
print(max_rating_realSum)
```

```python
average_rating = df['guest_satisfaction_overall'].mean()
print("Средний рейтинг удовлетворенности:", average_rating)
```

```python
average_rating = df['cleanliness_rating'].mean()
print("Средний рейтинг чистоты:", average_rating)
```

```python
profit_by_room_type = data.groupby('realSum')['person_capacity'].sum()
most_profitable_room_type = profit_by_room_type.idxmax()
```

```python
print(f"Самая прибыльная комната: {most_profitable_room_type}")
```

```python
popular_room_type = data['room_type'].mode().values[0]
```

```python
print(f"Самый популярный тип комнат: {popular_room_type}")
```

```python
average_rating = df['dist'].mean()
print("Средний рейтинг дистанции:", average_rating)
```

```python
average_rating = df['metro_dist'].mean()
print("Средний рейтинг расстояния метро:", average_rating)
```

```python
room_type = df['room_type'].value_counts()
print("Количество комнат по типам:")
print(room_type)
```

```python
max_rest_index_norm = df[df['rest_index_norm'] == df['rest_index_norm'].max()]
print("Наивысший рейтинг покоя:")
print(max_rest_index_norm)
```

```python
max_attr_index_norm = df[df['attr_index_norm'] == df['attr_index_norm'].max()]
print("Норма рейтинга индекса атрибута:")
print(max_attr_index_norm)
```

```python
average_rating = df['person_capacity'].mean()
print("Средний рейтинг вместимости:", average_rating)
```

```python
average_rating = df['realSum'].mean()
print("Средний рейтинг суммы:", average_rating)
```

```python
max_host_is_superhost = df[df['host_is_superhost'] == df['host_is_superhost'].max()]
print("Норма рейтинга хозяина:")
print(max_host_is_superhost)
```

```python
average_rating = df['biz'].mean()
print("Средний рейтинг бизнеса:", average_rating)
```
